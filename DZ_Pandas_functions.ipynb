Задание 1**
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
  - оценка 2 и меньше - низкий рейтинг
  - оценка 4 и меньше - средний рейтинг
  - оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
movies_data = pd.read_csv('/Users/lalitatsarkova/Desktop/git/netology--dmar/ml-latest-small/movies.csv')
movies_data[['movieId', 'title']].head()
```

```python
ratings_data = pd.read_csv('/Users/lalitatsarkova/Desktop/git/netology--dmar/ml-latest-small/ratings.csv')
ratings_data[['movieId','rating']].head()
```

```python
joined = ratings_data.merge(movies_data, on = 'movieId', how ='left')
joined[['movieId','rating','title']].head()
```

```python
def class_(row):
    if row['rating'] <= 2:
        return 'низкий'
    elif row['rating'] <= 4:
        return 'средний'
    elif row['rating'] > 4:
        return 'высокий'
```

```python
joined['class_'] = joined.apply(class_, axis=1)
joined.head(10)
```

**Задание 2**
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

Результат классификации запишите в отдельный столбец region.


```python
keywords_data = pd.read_csv('/Users/lalitatsarkova/Desktop/git/netology--dmar/Python_11_pandas_functions_groupby/keywords.csv')
```

```python
keywords_data.head()
```

```python
geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}
```

```python
d = {}

for region in geo_data:
    
    spisok_gorodov = geo_data[region]
    
    d.update({key:value for key,value in zip(spisok_gorodov,
                                [region for i in spisok_gorodov])})
```

```python
keywords_data['region'] = keywords_data['keyword'].map(d)
keywords_data['region'] = keywords_data['region'].fillna('undefined')
keywords_data.info()
keywords_data.head(10)
```
