````python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
````

````python
def p():
    inp = input('Введите номер документа: ')
    for elem in documents:
        if elem['number'] == inp:
            the_name = elem['name']
            return f'Владелец документа: {the_name}'
    return "Документ не найден в базе"
        
p()
````

````python
def l():
    for x in documents:
        type = x['type']
        number = x['number']
        name = x['name']
        g = 'type: {0}, number: {1}, name: {2}'.format(type, number, name)
        print(g)
    
l()
````

````python
def s():
    num = input('Введите номер документа:')
    for shelf in directories:
        if num in directories[shelf]:
            the_num = f'Документ хранится в полке: {shelf}'
            return the_num
    return "Документ не найден в базе"
s()
````

````python
def ads():
    number = input("Введите номер новой полки: ")
    if number in directories:
        print('Такая полка уже существует')
    else:
        directories[number] = []
        print('Полка добавлена. Текущий перечень полок:',directories.keys())
    

    
    
ads()
````

````python
def ds():
    unwanted_shelf = input('Введите номер полки: ')
    if unwanted_shelf in directories.keys() and len(directories[unwanted_shelf]) > 0:
        print("На полке есть документы, удалите их перед удалением полки.")
    elif unwanted_shelf not in directories.keys():
        print("Такой полки не существует.")
    elif unwanted_shelf in directories.keys() and len(directories[unwanted_shelf]) == 0:
        directories.pop(unwanted_shelf)
        print("Полка удалена.")
    

ds()
print("Текущий перечень полок:" , directories.keys())
````

````python
command = input("Введите название команды: ")
if command == 'p':
    p()
if command == 's':
    s()
if command == 'l':
    l()
if command == 'ads':
    ads()
if command == 'ds':
    ds()
else:
    print('Вы ввели некорректную команду, повторите ввод.')
````
